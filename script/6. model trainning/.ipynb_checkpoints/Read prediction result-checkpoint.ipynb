```python
import file_utilities as fu
import tifffile
import numpy as np

file_path = fu.choose_file()

image = tifffile.imread(file_path)


print("shape: {}".format(image.shape))
print("dtype: {}".format(image.dtype))
print("range: ({}, {})".format(np.min(image), np.max(image)))
```

```python
import scipy.ndimage
import napari
image[image == 2] = 0 # remove border



markers, num_features = scipy.ndimage.label(image)
# 过滤标签，只保留大小大于1000的标签
min_label_size = 1000
filtered_markers = np.zeros_like(markers)
unique_labels, label_counts = np.unique(markers, return_counts=True)
for label, count in zip(unique_labels, label_counts):
    if count >= min_label_size:
        filtered_markers[markers == label] = label

print(f'There are {len(np.unique(filtered_markers))} axons!')
# 创建Napari Viewer并添加图像和标签
viewer = napari.Viewer()
labels_layer = viewer.add_labels(filtered_markers, name='20 Filtered Labels (3D Watershed Segmentation)')

# 显示Napari Viewer
napari.run()
```
